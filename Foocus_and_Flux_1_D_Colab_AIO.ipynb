```python
# @title ##### **To create a ZIP archive**
# Importing required libraries
from zipfile import ZipFile
from IPython.display import HTML, display, Javascript
import os

# Function to create a ZIP archive with a progress bar
def create_zip_archive_with_progress(folder_path, output_file):
    try:
        if not output_file.endswith('.zip'):
            output_file += '.zip'

        file_list = []
        for foldername, subfolders, filenames in os.walk(folder_path):
            for filename in filenames:
                filepath = os.path.join(foldername, filename)
                file_list.append(filepath)

        num_files = len(file_list)

        if num_files == 0:
            print(f"The folder {folder_path} is empty. No files to archive.")
            return

        # Display progress bar with percentage text
        display(HTML('''
            <div style="display: flex; align-items: center;">
                <progress value="0" max="100" style="flex-grow: 1; margin-right: 10px;"></progress>
                <span id="progress-text">0%</span>
            </div>
        '''))

        with ZipFile(output_file, 'w') as archive:
            for i, filepath in enumerate(file_list):
                archive.write(filepath, os.path.relpath(filepath, folder_path))

                # Calculate progress
                progress = (i + 1) / num_files * 100

                # Update progress bar and percentage text using JavaScript
                display(Javascript(f'''
                    var progressValue = {progress};
                    var progressText = Math.floor(progressValue) + '%';
                    document.querySelector("progress").value = progressValue;
                    document.querySelector("#progress-text").textContent = progressText;
                '''))

        print(f"Contents of the folder {folder_path} have been successfully archived to {output_file}")
    except Exception as e:
        print(f"Failed to archive the folder: {str(e)}")

# Example usage:
folder_path = ''  # @param {'type': 'string'}
output_dir = ''  # @param {'type': 'string'}
zip_name = ''  # @param {'type': 'string'}

if not output_dir.endswith('/'):
    output_dir += '/'

if not zip_name:
    zip_name = os.path.basename(os.path.normpath(folder_path)) + ".zip"

output_file = os.path.join(output_dir, zip_name)
create_zip_archive_with_progress(folder_path, output_file)
```

```python
# @title ##### **To extract ZIP or RAR files**
# Importing required libraries
import os
from zipfile import ZipFile
import rarfile
from IPython.display import HTML, display, Javascript

# Function to extract ZIP or RAR files with progress bar
def unzip_file(file_path, extract_path, delete_after=False):
    try:
        if not file_path:
            print("No file path provided for extraction.")
            return

        if not os.path.exists(file_path):
            print(f"The file {file_path} was not found.")
            return

        if not os.path.exists(extract_path):
            os.makedirs(extract_path)

        # Check file type and extract accordingly
        if file_path.endswith(".zip"):
            with ZipFile(file_path, 'r') as zip_ref:
                file_list = zip_ref.namelist()
                num_files = len(file_list)

                if num_files == 0:
                    print(f"The archive {file_path} is empty.")
                    return

                # Display progress bar with percentage text
                display(HTML('''
                    <div style="display: flex; align-items: center;">
                        <progress value="0" max="100" style="flex-grow: 1; margin-right: 10px;"></progress>
                        <span id="progress-text">0%</span>
                    </div>
                '''))

                for i, file_name in enumerate(file_list):
                    zip_ref.extract(file_name, extract_path)

                    # Calculate progress
                    progress = (i + 1) / num_files * 100

                    # Update progress bar and percentage text using JavaScript
                    display(Javascript(f'''
                        var progressValue = {progress};
                        var progressText = Math.floor(progressValue) + '%';
                        document.querySelector("progress").value = progressValue;
                        document.querySelector("#progress-text").textContent = progressText;
                    '''))

                print(f"The file {file_path} has been successfully extracted to {extract_path}")

        elif file_path.endswith(".rar"):
            with rarfile.RarFile(file_path) as rar_ref:
                file_list = rar_ref.namelist()
                num_files = len(file_list)

                if num_files == 0:
                    print(f"The archive {file_path} is empty.")
                    return

                # Display progress bar with percentage text
                display(HTML('''
                    <div style="display: flex; align-items: center;">
                        <progress value="0" max="100" style="flex-grow: 1; margin-right: 10px;"></progress>
                        <span id="progress-text">0%</span>
                    </div>
                '''))

                for i, file_name in enumerate(file_list):
                    rar_ref.extract(file_name, extract_path)

                    # Calculate progress
                    progress = (i + 1) / num_files * 100

                    # Update progress bar and percentage text using JavaScript
                    display(Javascript(f'''
                        var progressValue = {progress};
                        var progressText = Math.floor(progressValue) + '%';
                        document.querySelector("progress").value = progressValue;
                        document.querySelector("#progress-text").textContent = progressText;
                    '''))

                print(f"The file {file_path} has been successfully extracted to {extract_path}")
        else:
            print("Unsupported file format. Only ZIP and RAR files are supported.")

        # Delete the archive file after extraction if the option is enabled
        if delete_after:
            os.remove(file_path)
            print(f"The archive file {file_path} has been deleted after extraction.")
    except Exception as e:
        print(f"Failed to extract the file: {str(e)}")

# Example usage:
file_path = ''  # @param {'type': 'string'}
extract_path = ''  # @param {'type': 'string'}
delete_archive_after_extract = True  # @param {type: 'boolean'}
unzip_file(file_path, extract_path, delete_archive_after_extract)
```

```python
import os
import pdfkit
import requests
from datetime import datetime
from bs4 import BeautifulSoup
from docx import Document
from docx.shared import Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.shared import Inches

# Buat folder untuk menyimpan hasil dan log
output_folder = 'convert_history_log'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Fungsi untuk konversi HTML ke PDF
def convert_html_to_pdf(input_html_path, output_pdf_name=None):
    if output_pdf_name is None:
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        output_pdf_name = f"converted_{timestamp}.pdf"

    output_pdf_path = os.path.join(output_folder, output_pdf_name)
    try:
        config = pdfkit.configuration(wkhtmltopdf='/usr/bin/wkhtmltopdf')
        options = {'enable-local-file-access': None}
        pdfkit.from_file(input_html_path, output_pdf_path, configuration=config, options=options)

        log_path = os.path.join(output_folder, "conversion_log.txt")
        with open(log_path, "a") as log_file:
            log_file.write(f"{datetime.now()} - Converted to PDF: {input_html_path} -> {output_pdf_path}\n")

        print(f"Successfully converted {input_html_path} to {output_pdf_path}")
    except Exception as e:
        print(f"Error converting {input_html_path} to PDF: {e}")

# Fungsi untuk konversi HTML ke DOCX
def convert_html_to_docx(input_html_path, output_docx_name=None):
    if output_docx_name is None:
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        output_docx_name = f"converted_{timestamp}.docx"

    output_docx_path = os.path.join(output_folder, output_docx_name)

    try:
        # Baca file HTML
        with open(input_html_path, "r", encoding="utf-8") as html_file:
            html_content = html_file.read()

        # Parse HTML dengan BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Buat dokumen DOCX
        doc = Document()

        # Iterasi melalui elemen-elemen HTML
        for element in soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'li', 'img', 'b', 'i', 'u']):
            if element.name.startswith('h'):  # Heading
                paragraph = doc.add_paragraph()
                run = paragraph.add_run(element.get_text(strip=True))
                run.bold = True
                paragraph.style = f'Heading {element.name[-1]}'

            elif element.name == 'p':  # Paragraph
                paragraph = doc.add_paragraph()
                for child in element.contents:
                    if child.name == 'b':  # Bold text
                        run = paragraph.add_run(child.get_text(strip=True))
                        run.bold = True
                    elif child.name == 'i':  # Italic text
                        run = paragraph.add_run(child.get_text(strip=True))
                        run.italic = True
                    elif child.name == 'u':  # Underlined text
                        run = paragraph.add_run(child.get_text(strip=True))
                        run.underline = True
                    else:  # Regular text
                        paragraph.add_run(str(child).strip())
                paragraph.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT

            elif element.name == 'li':  # List item
                doc.add_paragraph(element.get_text(strip=True), style='List Bullet')

            elif element.name == 'img':  # Image
                img_src = element.get('src')
                if img_src:
                    # Download the image if it's an external link
                    if img_src.startswith('http'):
                        img_data = requests.get(img_src).content
                        img_path = os.path.join(output_folder, "temp_image.jpg")
                        with open(img_path, 'wb') as img_file:
                            img_file.write(img_data)
                        doc.add_picture(img_path, width=Inches(4))  # Adjust size as needed
                        os.remove(img_path)  # Clean up temporary image
                    else:
                        # If the image is local, add it directly
                        img_path = os.path.join(os.path.dirname(input_html_path), img_src)
                        if os.path.exists(img_path):
                            doc.add_picture(img_path, width=Inches(4))  # Adjust size as needed

        # Simpan dokumen sebagai file DOCX
        doc.save(output_docx_path)

        # Simpan log
        log_path = os.path.join(output_folder, "conversion_log.txt")
        with open(log_path, "a") as log_file:
            log_file.write(f"{datetime.now()} - Converted to DOCX: {input_html_path} -> {output_docx_path}\n")

        print(f"Successfully converted {input_html_path} to {output_docx_path}")
    except Exception as e:
        print(f"Error converting {input_html_path} to DOCX: {e}")

input_html_path = "https://d6937ab66f5e0ba06a.gradio.live/file=/content/Fooocus/outputs/2025-01-19/log.html"  # @param {type: 'string'}

save_as_pdf = False # @param {type: 'boolean'}
save_as_docx = False # @param {type: 'boolean'}
save_as_both = False # @param {type: 'boolean'}

# Logika untuk menentukan konversi
if save_as_both:
    convert_html_to_pdf(input_html_path)
    convert_html_to_docx(input_html_path)
elif save_as_pdf:
    convert_html_to_pdf(input_html_path)
elif save_as_docx:
    convert_html_to_docx(input_html_path)
else:
    print("Tidak ada pilihan konversi yang diaktifkan. Aktifkan salah satu opsi.")
```
